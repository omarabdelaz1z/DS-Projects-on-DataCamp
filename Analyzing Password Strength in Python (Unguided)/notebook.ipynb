# Password Requirments:

1. Must be at least 10 characters in length
2. Must contain at least:
    - one lower case letter
    - one upper case letter
    - one numeric character
    - one non-alphanumeric character
3. Must not contain the phrase password (case insensitive)
4. Must not contain the user's first or last name, e.g., if the user's name is John Smith, then SmItH876! is not a valid password.

```python
def has_bad_password(row):
    isValid = True
    
    if len(row.password) < 10:
        isValid = False
    
    
    pattern = re.compile(r'((?=.*\d)(?=.*[A-Z])(?=.*[a-z])(?=.*\W).{10})')
    matched = re.findall(pattern, row.password)
    
    if (not matched):
        isValid = False
    
    temp = []
    for token in row.username.lower().split('.'):
        temp.append(token.lower() in row.password.lower())
        
    if(sum(temp)>0):
        isValid = False
        
    if (bool(re.search('password', row.password, flags = re.IGNORECASE))):
        isValid = False
        
    if(not isValid):
        return row.name
```

```python
bad_passwords_indices = logins.apply(has_bad_password, axis = 1).dropna().astype(int).values
len(bad_passwords_indices)
```

```python
bad_users = logins.loc[bad_passwords_indices, ]
bad_users
```
